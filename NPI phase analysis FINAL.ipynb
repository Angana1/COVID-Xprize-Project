```python
import numpy as np
import pandas as pd
```

```python
df_orig=pd.read_csv('OxCGRT_latest.csv')
df_orig.head()
```

```python
country='India'
phases=3
```

```python
#df_country is the dataframe from OxCGRT with information about a single country
df_grouped=df_orig.groupby('CountryName')
df_country=df_grouped.get_group(country)
df_country.head()
```

```python
new_data=[]
#this list will be finally printed onto csv file
```

```python
for i in range(phases):
    
    backslash = '/'
    filename = country + backslash + country + '.csv_split' + str(i) + '.csv'
    # df stores information of a particular phase from 22nd jan to 8th dec
    df = pd.read_csv(filename)
    
    start_date = df.Date[0]
    
    if i==(phases-1):
        end_date=20201130
    else:
        end_date = df.Date[df.shape[0] - 1]
        
    #print(end_date)
    #print(' ')
    
    # dfNPI stores info about the country from start date to end date of current phase
    dfNPI= df_country.loc[(df_country[df_country['Date'] == start_date].index[0]):(df_country[df_country['Date'] == end_date].index[0]), :]
    #print(dfNPI.tail())
    
    days=dfNPI.shape[0]
    
    #C3_2 = dfNPI[dfNPI['C3_Cancel public events'] == 2].shape[0]

        
    #if i==2:
        #print('cases:',C3_2)
        
        
    
    C1_0 = dfNPI[dfNPI['C1_School closing'] == 0].shape[0]
    C1_1 = dfNPI[dfNPI['C1_School closing'] == 1].shape[0]
    C1_2 = dfNPI[dfNPI['C1_School closing'] == 2].shape[0]
    C1_3 = dfNPI[dfNPI['C1_School closing'] == 3].shape[0]

    C2_0 = dfNPI[dfNPI['C2_Workplace closing'] == 0].shape[0]
    C2_1 = dfNPI[dfNPI['C2_Workplace closing'] == 1].shape[0]
    C2_2 = dfNPI[dfNPI['C2_Workplace closing'] == 2].shape[0]
    C2_3 = dfNPI[dfNPI['C2_Workplace closing'] == 3].shape[0]

    C3_0 = dfNPI[dfNPI['C3_Cancel public events'] == 0].shape[0]
    C3_1 = dfNPI[dfNPI['C3_Cancel public events'] == 1].shape[0]
    C3_2 = dfNPI[dfNPI['C3_Cancel public events'] == 2].shape[0]

    C4_0 = dfNPI[dfNPI['C4_Restrictions on gatherings'] == 0].shape[0]
    C4_1 = dfNPI[dfNPI['C4_Restrictions on gatherings'] == 1].shape[0]
    C4_2 = dfNPI[dfNPI['C4_Restrictions on gatherings'] == 2].shape[0]
    C4_3 = dfNPI[dfNPI['C4_Restrictions on gatherings'] == 3].shape[0]
    C4_4 = dfNPI[dfNPI['C4_Restrictions on gatherings'] == 4].shape[0]

    C5_0 = dfNPI[dfNPI['C5_Close public transport'] == 0].shape[0]
    C5_1 = dfNPI[dfNPI['C5_Close public transport'] == 1].shape[0]
    C5_2 = dfNPI[dfNPI['C5_Close public transport'] == 2].shape[0]

    C6_0 = dfNPI[dfNPI['C6_Stay at home requirements'] == 0].shape[0]
    C6_1 = dfNPI[dfNPI['C6_Stay at home requirements'] == 1].shape[0]
    C6_2 = dfNPI[dfNPI['C6_Stay at home requirements'] == 2].shape[0]
    C6_3 = dfNPI[dfNPI['C6_Stay at home requirements'] == 3].shape[0]

    C7_0 = dfNPI[dfNPI['C7_Restrictions on internal movement'] == 0].shape[0]
    C7_1 = dfNPI[dfNPI['C7_Restrictions on internal movement'] == 1].shape[0]
    C7_2 = dfNPI[dfNPI['C7_Restrictions on internal movement'] == 2].shape[0]

    C8_0 = dfNPI[dfNPI['C8_International travel controls'] == 0].shape[0]
    C8_1 = dfNPI[dfNPI['C8_International travel controls'] == 1].shape[0]
    C8_2 = dfNPI[dfNPI['C8_International travel controls'] == 2].shape[0]
    C8_3 = dfNPI[dfNPI['C8_International travel controls'] == 3].shape[0]
    C8_4 = dfNPI[dfNPI['C8_International travel controls'] == 4].shape[0]

    H1_0 = dfNPI[dfNPI['H1_Public information campaigns'] == 0].shape[0]
    H1_1 = dfNPI[dfNPI['H1_Public information campaigns'] == 1].shape[0]
    H1_2 = dfNPI[dfNPI['H1_Public information campaigns'] == 2].shape[0]

    H2_0 = dfNPI[dfNPI['H2_Testing policy'] == 0].shape[0]
    H2_1 = dfNPI[dfNPI['H2_Testing policy'] == 1].shape[0]
    H2_2 = dfNPI[dfNPI['H2_Testing policy'] == 2].shape[0]
    H2_3 = dfNPI[dfNPI['H2_Testing policy'] == 3].shape[0]

    H3_0 = dfNPI[dfNPI['H3_Contact tracing'] == 0].shape[0]
    H3_1 = dfNPI[dfNPI['H3_Contact tracing'] == 1].shape[0]
    H3_2 = dfNPI[dfNPI['H3_Contact tracing'] == 2].shape[0]

    H6_0 = dfNPI[dfNPI['H6_Facial Coverings'] == 0].shape[0]
    H6_1 = dfNPI[dfNPI['H6_Facial Coverings'] == 1].shape[0]
    H6_2 = dfNPI[dfNPI['H6_Facial Coverings'] == 2].shape[0]
    H6_3 = dfNPI[dfNPI['H6_Facial Coverings'] == 3].shape[0]
    H6_4 = dfNPI[dfNPI['H6_Facial Coverings'] == 4].shape[0]


    # function to calculate percentages
    def perc(num, days):
        return ((num / days) * 100)


    dict_rows = {}

    dict_rows.update({'Phase': i})
    dict_rows.update({'C1_School closing 0': perc(C1_0, days)})
    dict_rows.update({'C1_School closing 1': perc(C1_1, days)})
    dict_rows.update({'C1_School closing 2': perc(C1_2, days)})
    dict_rows.update({'C1_School closing 3': perc(C1_3, days)})

    dict_rows.update({'C2_Workplace closing 0': perc(C2_0, days)})
    dict_rows.update({'C2_Workplace closing 1': perc(C2_1, days)})
    dict_rows.update({'C2_Workplace closing 2': perc(C2_2, days)})
    dict_rows.update({'C2_Workplace closing 3': perc(C2_3, days)})

    dict_rows.update({'C3_Cancel public events 0': perc(C3_0, days)})
    dict_rows.update({'C3_Cancel public events 1': perc(C3_1, days)})
    dict_rows.update({'C3_Cancel public events 2': perc(C3_2, days)})

    dict_rows.update({'C4_Restrictions on gatherings 0': perc(C4_0, days)})
    dict_rows.update({'C4_Restrictions on gatherings 1': perc(C4_1, days)})
    dict_rows.update({'C4_Restrictions on gatherings 2': perc(C4_2, days)})
    dict_rows.update({'C4_Restrictions on gatherings 3': perc(C4_3, days)})
    dict_rows.update({'C4_Restrictions on gatherings 4': perc(C4_4, days)})

    dict_rows.update({'C5_Close public transport 0': perc(C5_0, days)})
    dict_rows.update({'C5_Close public transport 1': perc(C5_1, days)})
    dict_rows.update({'C5_Close public transport 2': perc(C5_2, days)})

    dict_rows.update({'C6_Stay at home requirements 0': perc(C6_0, days)})
    dict_rows.update({'C6_Stay at home requirements 1': perc(C6_1, days)})
    dict_rows.update({'C6_Stay at home requirements 2': perc(C6_2, days)})
    dict_rows.update({'C6_Stay at home requirements 3': perc(C6_3, days)})

    dict_rows.update({'C7_Restrictions on internal movement 0': perc(C7_0, days)})
    dict_rows.update({'C7_Restrictions on internal movement 1': perc(C7_1, days)})
    dict_rows.update({'C7_Restrictions on internal movement 2': perc(C7_2, days)})

    dict_rows.update({'C8_International travel controls 0': perc(C8_0, days)})
    dict_rows.update({'C8_International travel controls 1': perc(C8_1, days)})
    dict_rows.update({'C8_International travel controls 2': perc(C8_2, days)})
    dict_rows.update({'C8_International travel controls 3': perc(C8_3, days)})
    dict_rows.update({'C8_International travel controls 4': perc(C8_4, days)})

    dict_rows.update({'H1_Public information campaigns 0': perc(H1_0, days)})
    dict_rows.update({'H1_Public information campaigns 1': perc(H1_1, days)})
    dict_rows.update({'H1_Public information campaigns 2': perc(H1_2, days)})

    dict_rows.update({'H2_Testing policy 0': perc(H2_0, days)})
    dict_rows.update({'H2_Testing policy 1': perc(H2_1, days)})
    dict_rows.update({'H2_Testing policy 2': perc(H2_2, days)})
    dict_rows.update({'H2_Testing policy 3': perc(H2_3, days)})

    dict_rows.update({'H3_Contact tracing 0': perc(H3_0, days)})
    dict_rows.update({'H3_Contact tracing 1': perc(H3_1, days)})
    dict_rows.update({'H3_Contact tracing 2': perc(H3_2, days)})

    dict_rows.update({'H6_Facial Coverings 0': perc(H6_0, days)})
    dict_rows.update({'H6_Facial Coverings 1': perc(H6_1, days)})
    dict_rows.update({'H6_Facial Coverings 2': perc(H6_2, days)})
    dict_rows.update({'H6_Facial Coverings 3': perc(H6_3, days)})
    dict_rows.update({'H6_Facial Coverings 4': perc(H6_4, days)})

    new_data.append(dict_rows)  # keep appending list
    #print(new_data)  # print list
    
    

    #dict_rows.clear() #clear the dictionary for next phase's iteration
```

```python
dataframe = pd.DataFrame(new_data)
dataframe.to_csv(country+' NPI phase analysis part 1.csv', index=False)
```
